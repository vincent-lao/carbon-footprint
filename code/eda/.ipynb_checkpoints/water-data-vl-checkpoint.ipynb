```python
sm_ghg = pd.read_csv('../../data/RICAPS_Water_Contribution_to_Greenhouse_Gas_Emissions.csv')
sm_ghg.head()
```

```python
coords = sm_ghg['Location 1'].apply(lambda s: np.array(re.findall('(-?\d{2,3}.\d+)', s), dtype=np.float))
sm_ghg.loc[:, 'Latitude'] = coords.map(lambda x: x[0])
sm_ghg.loc[:, 'Longitude'] = coords.map(lambda x: x[1])
sm_ghg.head()
```

```python
from urllib.request import urlopen
import json
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/fips-unemp-16.csv",
                   dtype={"fips": str})

import plotly.express as px

fig = px.choropleth(df, geojson=counties, locations='fips', color='unemp',
                           color_continuous_scale="Viridis",
                           range_color=(0, 12),
                           scope="california",
                           labels={'unemp':'unemployment rate'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
