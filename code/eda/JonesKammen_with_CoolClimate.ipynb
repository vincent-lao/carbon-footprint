<h3>Concatenated CoolClimate Datasets</h3>

```python
coolClimate = pd.concat([coolClimate_household_1, coolClimate_household_2, coolClimate_household_3, coolClimate_household_4,
                        coolClimate_household_5])
coolClimate.head()
coolClimate.shape
```

```python
coolClimate['Zip Code'] = coolClimate['Zip Code'].astype('int')

coolClimate['Zip Code']

coolClimate = coolClimate.rename(columns={'Zip Code': 'ZipCode'})

coolClimate.head()
```

```python
coolClimate_grouped_by = coolClimate.groupby(['ZipCode']).sum()
coolClimate_grouped_by.head()
```

```python
jonesKammen_ca_zipcode = jonesKammen_ca_zipcode.drop(columns=["Latitude", "Longitude", "Elevation", "State", "StateFullName", 
                                    "City", "CBSAType", "CBSAName","MSAName", "PMSA_Name", "Region", 
                                    "Division", "Transport (tCO2e/yr)", "Housing (tCO2e/yr)", "Food (tCO2e/yr)",
                                    "Goods (tCO2e/yr)", "Services (tCO2e/yr)", 
                                     "Total Household Carbon Footprint (tCO2e/yr)", "Total Zip Code Carbon Footprint (tCO2e/yr)"])
jonesKammen_ca_zipcode.head()

```

```python
mergedZC = coolClimate_grouped_by.join(jonesKammen_ca_zipcode.set_index('ZipCode'))
mergedZC
```

```python
final_dataset = mergedZC.groupby(['CountyName']).sum()
final_dataset = final_dataset.drop(columns=["input_income", 
                                            'input_footprint_transportation_mpg1',
                                            'input_footprint_transportation_fuel1',
                                            'input_footprint_housing_cdd',
                                            'input_footprint_housing_hdd',
                                            'input_footprint_housing_watersewage',
                                            'input_footprint_shopping_goods_other_personalcare',
                                            'input_footprint_shopping_services_education',
                                            'input_footprint_shopping_services_total',
                                            'input_footprint_shopping_food_meatfisheggs_default'
                                           ])
```

```python
final_dataset = final_dataset.rename({'input_footprint_transportation_num_vehicles': 'num_vehicles',
                                     'input_footprint_transportation_miles1': 'miles_traveled_year',
                                     'input_footprint_transportation_airshort': 'num_short_flights',
                                     'input_footprint_transportation_airmedium': 'num_med_flights',
                                     'input_footprint_transportation_airlong': 'num_long_flights',
                                     'input_footprint_housing_electricity_dollars': 'electricity_billPerYear',
                                      'input_footprint_housing_squarefeet': 'housing_squarefeet',
                                      'input_footprint_shopping_goods_clothing': 'moneySpentOnClothesPerYear'
                                     },
                                     
                                     axis = 1)


final_dataset
```

```python
final_dataset.to_csv('../../data/jonesKammen-PlUS-CoolClimate.csv')
```
